```
from PIL import Image
import pylab
import matplotlib.pyplot as plt

im1 = np.array(Image.open('/content/gdrive/My Drive/ChestAI/48011458/48011458_CR_20060206_001crop.png'))
im2 = np.array(Image.open('/content/gdrive/My Drive/ChestAI/48011458/48011458_CR_20190121_001crop.png'))
im3 = np.array(Image.open('/content/gdrive/My Drive/ChestAI/48011458/48011458_CR_20190318_001crop.png'))

fig = plt.figure(figsize=(15, 25))

plt.subplot(1,3,1), plt.imshow(im1,cmap='bone'),plt.title('20060206')
plt.subplot(1,3,2), plt.imshow(im2,cmap='bone'),plt.title('20190121')
plt.subplot(1,3,3), plt.imshow(im3,cmap='bone'),plt.title('20190318');
```

```
params = ['none', 'nearest', 'bilinear', 'bicubic', 'spline16', 'spline36',
          'hanning', 'hamming', 'hermite', 'kaiser', 'quadric', 'catrom',
          'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos']

fig = plt.figure(figsize=(12, 20))
for i, p in enumerate(params, 1):
    ax = fig.add_subplot(6, 3, i)
    ax.imshow(im1, interpolation=p)
    ax.set_axis_off()
    
    s = "'{}'".format(p) if isinstance(p, str) else p
    ax.set_title("interpolation={}".format(s))
    
plt.show()
```

```
im111 = plt.imread('/content/gdrive/My Drive/ChestAI/48011458/20060206-1edge.png').astype(float)


from scipy import fftpack
im111_fft = fftpack.fft2(im111)

# Show the results

def plot_spectrum(im111_fft):
    from matplotlib.colors import LogNorm
    # A logarithmic colormap
    plt.imshow(np.abs(im111_fft), norm=LogNorm(vmin=5))
    plt.colorbar()

plt.figure()
plot_spectrum(im111_fft)
plt.title('20060206 FFT')
```

```
im211 = plt.imread('/content/gdrive/My Drive/ChestAI/48011458/20190121edge.png').astype(float)


from scipy import fftpack
im211_fft = fftpack.fft2(im211)

# Show the results

def plot_spectrum(im211_fft):
    from matplotlib.colors import LogNorm
    # A logarithmic colormap
    plt.imshow(np.abs(im211_fft), norm=LogNorm(vmin=5))
    plt.colorbar()

plt.figure()
plot_spectrum(im211_fft)
plt.title('20190121FFT')
```
